```python
import json
from tqdm import tqdm
filename = "./data/CDD/500/json/train.json"
sum = 0
sentences = []

with open("./data/CDD/500/conll/train.txt", "w", encoding="utf-8") as out:
    with open(filename, "r", encoding="utf-8") as f:
        for line in tqdm(f):
            data = json.loads(line)
            text = data["text"]
            label = data["label"]
            for i in range(len(text)):
                out.write(text[i] + "	" + label[i] + "\n")
            out.write('\n')

```
